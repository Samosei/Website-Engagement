```python
import pandas as pd

df = pd.read_csv('gov-website-metrics-by-day.csv')

df
```

```python
# Checking if there are any nulls
df.isnull().values.any()
```

```python
df.dtypes
```

```python
df["Date"] = pd.to_datetime(df["Date"]).dt.strftime('%Y-%m-%d')
df
```

```python
# filtering out needed coluumns for analysis
df = df.filter(['Date','Visits','Unique Visitors','Actions','Maximum actions in one visit','Bounces','Total time spent by visitors (in seconds)','Pageviews','Unique Pageviews','Downloads','Searches','Unique Keywords'], axis = 1)
df
```

```python
df['Engagement'] = df['Visits']/df['Total time spent by visitors (in seconds)']
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
plt.plot(df['Date'], df['Total time spent by visitors (in seconds)'],color='blue', marker='o')
plt.title('Daily Engagement by Total Time Spent By Visitors for Whole Period', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Total time spent by visitors (in seconds)')


plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
plt.plot(df['Date'], df['Visits'],color='navy', marker='o')
plt.title('Daily Engagement by Visits for Whole Period', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Number of Visits')


plt.show()
```

```python
# Filter data between two dates for phase one - 151 days
df1 = df.loc[(df['Date'] >= '2016-10-01')
                     & (df['Date'] <= '2017-02-28')]
# Display
df1
```

```python
mean_visits_one = df1['Visits'].mean()
mean_time_one = df1['Total time spent by visitors (in seconds)'].mean()
print('Average number of visits for phase one', mean_visits_one )
print('Total time spent by visitors (in seconds) for phase one',mean_time_one )
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))

plt.plot(df1['Date'], df1['Visits'],color='navy', marker='o')

plt.title('Daily Engagement by Visits Phase One', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Number of Visits')


plt.axhline(y=mean_visits_one , color='green', linestyle='-')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
plt.plot(df1['Date'], df1['Total time spent by visitors (in seconds)'],color='blue', marker='o')
plt.title('Daily Engagement by Total Time Spent By Visitors for Phase One', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Total time spent by visitors (in seconds)')


plt.axhline(y=mean_time_one , color='green', linestyle='-')

plt.show()
```

```python
# Filter data between two dates for phase two - 246 days
df2 = df.loc[(df['Date'] > '2017-02-28')]
# Display
df2
```

```python
mean_visits_two = df2['Visits'].mean()
mean_time_two = df2['Total time spent by visitors (in seconds)'].mean()
print('Average number of visits for phase two', mean_visits_two)
print('Total time spent by visitors (in seconds) for phase two',mean_time_two)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
plt.plot(df2['Date'], df2['Total time spent by visitors (in seconds)'],color='blue', marker='o')
plt.title('Daily Engagement by Total Time Spent By Visitors for Phase Two', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Total time spent by visitors (in seconds)')

plt.axhline(y=mean_time_two, color='orange', linestyle='-')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
plt.plot(df2['Date'], df2['Visits'],color='navy', marker='o')
plt.title('Daily Engagement by Visits Phase Two', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Number of Visits')

plt.axhline(y=mean_visits_two , color='orange', linestyle='-')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats

# Extract the day of the week from the date column
df['Day_of_Week'] = pd.to_datetime(df['Date']).dt.dayofweek


# Map the integer values to the corresponding day of the week
df['Day_of_Week'] = df['Day_of_Week'].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})


average_number_of_visitors = df["Visits"].mean()


# Group the data by day of the week and calculate the average wait time for each group
grouped_data = df.groupby('Day_of_Week').mean()

# Create a bar plot to compare the average wait time for each day of the week to the overall average wait time

sns.set(rc={'figure.figsize':(15,10)})

sns.barplot(x=grouped_data.index, y='Visits', data=grouped_data)


plt.ylabel('Average No. of Visits')
plt.title('Average Number of Vists by Day of the Week')

# Show the plot
plt.show()

grouped_data['Visits']
```

Tuesday, Monday, Wednesday , and Thursday in decreasing order have the highest traffic on website by using total number of visitors. 

There is a drop in thge number of visitors on Friday which which is the last typical working day of the weekday and the weekends have the least number of visits.
